```python
# tuple: cannot append, remove or change elements
# tuple: immutable lists

tup1 = ( 1, 2, 3)

print(tup1)
print(tup1[0])

tup1[0] = 10 # TypeError: 'tuple' object does not support item assignment
print(tup1)
```

```python
# list - content copying and not references

list_1 = [ 1, 2, 3]
list_2 = list_1[:]

print("{}\n{}".format(list_1,list_2))

list_1[0] = "lokeshwar"

print()
print("{}\n{}".format(list_1,list_2))

# since changes are made to the reference list_1, therefor, list_2 is NOT modified, as it is a COPY
```

```python
# lists - references concept
list_1 = [ 1, 2, 3]
list_2 = list_1

print("{}\n{}".format(list_1,list_2))

list_1[0] = "lokeshwar"

print()
print("{}\n{}".format(list_1,list_2))

# since changes are made to the reference list_1, therefor, list_2 is also modified
```

```python
# lists
l1 = [ 'a', 1, True ]

print("type() = ", type(l1))
print("list values = ", l1)
print("length of list = ", len(l1))

# append in list
l1.append("lokeshwar")
print("list after appending", l1)

# extending an list to another list
l2 = [ "hello", "world", False]
print("l2 = ", l2)
l2.extend(l1)
print()
print("after extending, l1 = ", l1)
print("after extending, l2 = ", l2)

# pop() operation on list
print()
print("pop = ", l2.pop())
print("list after pop() = " , l2)

# pop(index)
print()
popIdx = 3
print("pop({}) =".format(popIdx),l2.pop(popIdx))
print("list after pop({}) =".format(popIdx), l2)
```
