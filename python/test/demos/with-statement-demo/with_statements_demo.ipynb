```python
# file handling w/o with and try..except statements
file_path = 'F:/git/tutorials/python/test/demos/with-statement-demo/demo.txt'
file = open(file_path,'w')
file.write('hello world!')
file.close()
```

```python
file_path = 'F:/git/tutorials/python/test/demos/with-statement-demo/demo2.txt'
try:
    file2 = open(file_path,'w')
    file2.write('hello world 2!')
    print("data written to file")
except Exception as err:
    # print('caught exception as -', ex, err)
    # traceback.print_exception(None, err, err.__traceback__)
    tb = traceback.TracebackException.from_exception(err)
    print(''.join(tb.stack.format()))
finally:
    file2.close()
print("program finished")    
```
