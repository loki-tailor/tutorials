```python
# ref: https://www.geeksforgeeks.org/iterators-in-python/

# custom iterator

class Test:

    # constructor
    def __init__(self, limit):
        self.limit = limit

    # creates iterator object
    # called when iteration is initialized
    def __iter__(self):
        self.x = 10
        return self

    # to move to next element
    def __next__(self):
    
        # store current value
        x = self.x

        # stop iteration if limit reached
        if x > self.limit:
            raise StopIteration

        # increment & return old value
        self.x += 1
        return x

# class-end

for i in Test(15):
    print(i, end = " ")

print()
for i in Test(5):
    print(i) # won't print 
    
```

```python
# custom class __iter__ & __next__
class CustNos:
    def __iter__(self):
        self.a = 1
        return self
    
    def __next__(self):
        if self.a <= 20:
            x = self.a
            self.a += 1
            return x
        else:
            raise StopIteration

c = CustNos()
itrx = iter(c)
print("--------------------------------")
print(c)
print(itr)
print(type(c))
print(type(itr))
print("--------------------------------")
for xx in itr:
    print(next(itrx), end = " ") # this loops until StopIteration is raised
print("\n--------------------------------")
itr2 = iter(c)
for i in range(21):
    print(next(itr2), end = " ") # this one throws exception
```
