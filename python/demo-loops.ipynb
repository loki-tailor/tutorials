```python
# while loop demo

i = 10

while i > 0:
    print(i)
    i -= 1

print(i)    
```

```python
# for with enumerate

list_1 = [ "loki", "thor", "captian america", "krish"]

for i, val in enumerate(list_1):
    print("index = {}, value = {}".format(i,val))
print()    
```

```python
# for demo with list

for i in [1, 2, "lokeshwar"]:
    print(i)
print()

# ----------------------------
list_1 = [ "loki", True, "and Thor, ", "hahahah"]
for i in range(len(list_1)):
    print(list_1[i])
```

```python
# range e.g.

for i in range(5):
    print(i)
print()

for i in range(10, 0, -1):
    print(i)

###
# range(start, stop, hops)
# e.g.  
###
```
